1. Setup and Imports

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import os
```

2- Create Output Folder

```python
os.makedirs("output_API", exist_ok=True)
```

3. Make API Request

```python
api_url = "https://en.wikipedia.org/api/rest_v1/page/summary/Cyrus_Cylinder"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36'}
response = requests.get(api_url, headers=headers)
response.status_code
```

4. Parse JSON Response

```python
data = response.json()
data.keys()
```

5. Convert API Data to Pandas DataFrame

```python
api_df = pd.DataFrame([{
    "title": data.get("title"),
    "description": data.get("description"),
    "summary": data.get("extract"),
    "language": data.get("lang"),
    "page_id": data.get("pageid"),
    "timestamp": data.get("timestamp")
}])

api_df
```

6. Export API Results to CSV

```python
api_df.to_csv("output_API/cyrus_cylinder_api_data.csv", index=False)
```
